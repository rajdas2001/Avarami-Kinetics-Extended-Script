```python
import numpy as np
import math 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x = int(input("Enter the number of iterations: "))
ns = int(input("Enter the number of nucleating sites: "))
tt = int(input("Enter the transformation temperature: "))
tu = int(input("Enter the undercooling temperature: "))
t1 = np.linspace(tu,tt,x)

print(t1)

ar = []
for elements in t1:
  ar.append(tt - elements)

print(ar)

final = []

for elements in ar:
  y = 1 - math.exp(-1766.25 * (10 ** -18) * ns * (elements ** 3) * (10000 ** 3) )
  final.append(y)

print(final)

plt.plot(ar, final)
plt.show()



res = []

for elements in ar:
  for i in range(1000,20001,180):
    y2 = 1 - math.exp(-1766.25 * (10 ** -18) * ns * (elements ** 3) * (i ** 3) )
    res.append([elements, i, y2])

print(res)

mat1 = np.array(res)
x1 = mat1[:, 0]
y1 = mat1[:, 1]
z1 = mat1[:, 2]

print(x1)
print(y1)
print(z1)

print(len(x1))
print(len(y1))
print(len(z1))

fig = plt.figure()
ax = plt.axes(projection ='3d')
surf = ax.plot_trisurf(x1, y1, z1, linewidth=0, antialiased=False)
ax.set_title('Surface plot')
plt.show()

```
